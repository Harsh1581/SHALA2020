## Part 2 - Drawing correlation and heatmap
- Using the correlation matrix, peform the following:
- Find the correlation matrix for the movies.
- Draw a heatmap with the correlation matrix and display the values of correlation coefficients for each pair of features.
- Create a mask to remove the duplicate values from the heatmap.

```python
corr_values = movies.corr(numeric_only = True)
plt.figure(figsize = (15,10))
sns.heatmap(corr_values, vmin = -1, cmap = 'coolwarm', annot = True)
plt.show()
```

```python
mask = np.triu(np.ones_like(corr_values, dtype=bool))

plt.figure(figsize = (15,10))
sns.heatmap(corr_values, vmin = -1, cmap = 'coolwarm', annot = True, mask = mask)
plt.show()
```

### Using a scatter plot, perform the following:

- Draw a scatter plot between between petal_length and petal_width and color it according to the species.
- Change the size of marker to 100 and place the legend in lower right.

Using density plots, perform the following:
- Load the dataset iris. This has been done for you.
- Draw the density plots for all the features sepal_width, sepal_length, petal_width, and petal_length.
- Try drawing a CDF of these features. It will be covered in tutorial.

```python
from pylab import *

subplot(2,2,1)
sns.kdeplot(data = df.sepal_width, shade =False, color = 'r', fill = True)

subplot(2,2,2)
sns.kdeplot(data = df.sepal_length, shade =True, color = 'b')

subplot(2,2,3)
sns.kdeplot(data = df.petal_width, shade =True, color = 'g')

subplot(2,2,4)
sns.kdeplot(data = df.petal_length, shade =True, color = 'y')
```

```python
from pylab import *

subplot(2,2,1)
sns.kdeplot(data = df.sepal_width, shade =True, color = 'r', cumulative = True, legend = False)

subplot(2,2,2)
sns.kdeplot(data = df.sepal_length, shade =True, color = 'b',  cumulative = True, legend = True)


subplot(2,2,3)
sns.kdeplot(data = df.petal_width, shade =True, color = 'g',  cumulative = True, legend = True)

subplot(2,2,4)
sns.kdeplot(data = df.petal_length, shade =True, color = 'y',  cumulative = True, legend = True)
```
